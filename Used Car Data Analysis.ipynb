# Which Fuel type is preferred the most?

```python
plt.figure(figsize=(10,10))
plt.title("Preferred Fuel Type", color = 'k', fontsize = 15)
df['Fuel_type'].value_counts().plot(kind='pie', autopct='%1.1f%%',shadow=True, cmap='viridis', legend=True)
```

### Petrol is the most preferred Fuel type, followed by Diesel. 

```python
sns.countplot(df['Fuel_type'], order = df['Fuel_type'].value_counts().iloc[:10].index, hue = df['Fuel_type'])
plt.title('Car\'s Fuel Type')
plt.legend(['Petrol', 'Diesel'])
```
